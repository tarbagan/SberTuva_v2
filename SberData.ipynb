```python
#autor: Irgit Valery
#Анализ открытых данных Сбербанка
import pandas as pd
import matplotlib.pyplot as plt
```

```python
URL = 'http://sbrfdata.ru/opendata.zip'
df = pd.read_csv(URL, compression='zip', sep=',', header=0, encoding='cp1251')
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
#Средняя зарплата по регионам
tuva = df.loc[df['region'] == 'Республика Тыва']
tuva = tuva.loc[df['name'] == 'Средняя зарплата']
hakas = df.loc[df['region'] == 'Республика Хакасия']
hakas = hakas.loc[df['name'] == 'Средняя зарплата']
kras = df.loc[df['region'] == 'Красноярский край']
kras = kras.loc[df['name'] == 'Средняя зарплата']

plt.figure(figsize=(13,7))
plt.title('Средняя зарплата по регионам')
plt.ylabel('Сумма')
plt.plot(tuva['date'], tuva['value'],label='Республика Тыва')
plt.plot(hakas['date'], hakas['value'],label='Республика Хакасия')
plt.plot(kras['date'], kras['value'],label='Красноярский край')
plt.legend()
```

```python
#кредитов на кол-во населения
tuva = df.loc[df['region'] == 'Республика Тыва']
tuva = tuva.loc[df['name'] == 'Количество заявок на потребительские кредиты']
hakas = df.loc[df['region'] == 'Республика Хакасия']
hakas = hakas.loc[df['name'] == 'Количество заявок на потребительские кредиты']
kras = df.loc[df['region'] == 'Красноярский край']
kras = kras.loc[df['name'] == 'Количество заявок на потребительские кредиты']

plt.figure(figsize=(13,7))
plt.title('Потребительских кредитов на кол-во населения')
plt.ylabel('Количество кредитов')
plt.plot(tuva['date'], tuva['value']/324420,label='Республика Тыва')
plt.plot(hakas['date'], hakas['value']/536167,label='Республика Хакасия')
plt.plot(kras['date'], kras['value']/2874050,label='Красноярский край')
plt.legend()
```

```python
#суммы кредита
df = df.loc[df['region'] == 'Республика Тыва']
df = df.loc[df['name'] == 'Средняя сумма заявки на потребительский кредит']
plt.figure(figsize=(13,7))
plt.title("Средняя сумма заявки на потребительский кредит Сбербанк-Тыва. Всего -%s рублей" % str(df['value'].sum()))
plt.ylabel('Суммы кредитов')
plt.plot(df['date'], df['value'])
```

```python
#всего кредитов
df = df.loc[df['region'] == 'Республика Тыва']
df = df.loc[df['name'] == 'Количество заявок на потребительские кредиты']
plt.figure(figsize=(13,7))
plt.title("Выдано потребительских кредитов Сбербанк-Тыва. Всего-%s" % str(df['value'].sum()))
plt.ylabel('Количество кредитов')
plt.plot(df['date'], df['value'])
```

```python
#ипотека
df = df.loc[df['region'] == 'Республика Тыва']
df = df.loc[df['name'] == 'Средняя сумма заявки на ипотечный кредит']
plt.figure(figsize=(13,7))
plt.title("Суммы на ипотечные кредиты Сбербанк-Тыва. Всего-%s рублей" % str(df['value'].sum()))
plt.ylabel('Сумма')
plt.plot(df['date'], df['value'])
```

```python
#ипотека
df = df.loc[df['region'] == 'Республика Тыва']
df = df.loc[df['name'] == 'Средняя сумма заявки на ипотечный кредит']
plt.figure(figsize=(13,7))
plt.title("Квадратных метров. Всего-%s" % str(df['value'].sum()))
plt.ylabel('Количество кредитов')
plt.bar(df['date'], df['value']/54000, width=15)
```

```python
df = df.loc[df['region'] == 'Республика Тыва']
df = df.loc[df['name'] == 'Средняя пенсия']
me = df.groupby(df['date'].dt.year)['value'].agg(['min', 'mean', 'max'])
me
plt.figure(figsize=(13,7))
plt.title('Средняя пенсия в Туве')
me['mean'].plot.bar()
```

```python
#Средний чек в формате Ресторан
tuva = df.loc[df['region'] == 'Республика Тыва']
tuva = tuva.loc[df['name'] == 'Средний чек в формате Ресторан']
hakas = df.loc[df['region'] == 'Республика Хакасия']
hakas = hakas.loc[df['name'] == 'Средний чек в формате Ресторан']
kras = df.loc[df['region'] == 'Красноярский край']
kras = kras.loc[df['name'] == 'Средний чек в формате Ресторан']

plt.figure(figsize=(13,7))
plt.title('Средний чек в формате Ресторан')
plt.ylabel('Сумма')
plt.plot(tuva['date'], tuva['value'],label='Республика Тыва')
plt.plot(hakas['date'], hakas['value'],label='Республика Хакасия')
plt.plot(kras['date'], kras['value'],label='Красноярский край')
plt.legend()
```
